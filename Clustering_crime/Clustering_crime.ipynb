```python
# Import libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch 
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import AgglomerativeClustering
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('crime_data.csv')
df
```

```python
# Data preprocessing and EDA
```

```python
df = df.rename({'Unnamed: 0':'State'}, axis = 1)
```

```python
df.info()
```

# Agglomerative clustering

```python
# Normalizing dataset

scalar = MinMaxScaler()
scalar_df = scalar.fit_transform(df.iloc[:,1:])
scalar_df
```

```python
# Create dendograms

plt.rcParams['figure.figsize'] = (10,6)
plt.rcParams['figure.dpi'] = 250

dendo = sch.dendrogram(sch.linkage(scalar_df, 'complete'))
```

```python
# Creating Clusters

H_cluster = AgglomerativeClustering(n_clusters = 4, affinity = 'euclidean', linkage = 'ward')
H_cluster
```

```python
# Using data normalized by MinMaxScaler 

y = pd.DataFrame(H_cluster.fit_predict(scalar_df), columns = ['Clustersid'])
y['Clustersid'].value_counts()
```

```python
# Adding clusters to dataset

df['Clustersid_HC'] = H_cluster.labels_
df
```

```python
df.groupby('Clustersid_HC').agg(['mean']).reset_index()
```

```python
# Plotting barplot using groupby method 
```

```python
df.groupby(['Clustersid_HC']).count()['State'].plot(kind='bar')
plt.title('Hirarchical Clustering')
plt.xlabel('Clusters')
plt.ylabel('States')
```

```python
# silhouette score of AgglomerativeClustering

from sklearn.metrics import silhouette_score
```

```python
ss = silhouette_score(scalar_df, H_cluster.labels_)
ss
```

```python
# States in cluster 0 

df[df['Clustersid_HC'] == 0]
```

# K-Means Clustering

```python
# Import Library
from sklearn.cluster import KMeans
```

```python
# Using data normalized by MinMaxScaler

wcss = []
for i in range(1,11):
    km = KMeans(n_clusters = i, random_state = 0)
    km.fit(scalar_df)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid()
```

```python
# Build cluster algorithm

km_cluster =  KMeans(4, random_state = 42)
km_cluster.fit(scalar_df)
```

```python
y = pd.DataFrame(km_cluster.fit_predict(scalar_df),columns = ['Clusterid_Kmeans'])
y['Clusterid_Kmeans'].value_counts()
```

```python
#Assign clusters to the data set

df['Clusterid_Kmeans'] = km_cluster.labels_
df
```

```python
df.groupby('Clusterid_Kmeans').agg(['mean']).reset_index()
```

```python
# Plotting barplot using groupby method
```

```python
df.groupby(['Clusterid_Kmeans']).count()['State'].plot(kind='bar')
plt.title('KMeans Clustering')
plt.xlabel('Clusters')
plt.ylabel('States')
```

```python
# States in cluster 0 

df[df['Clusterid_Kmeans'] == 0]
```

```python
# States in cluster 1
df[df['Clusterid_Kmeans'] == 1]
```

```python
# States in cluster 2

df[df['Clusterid_Kmeans'] == 2]
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
EPS = [0.22, 0.24, 0.26,0.28, 0.30]

for n in EPS:
    dbscan = DBSCAN(eps=n, min_samples=4)
    dbscan.fit(scalar_df)
    y = pd.DataFrame(dbscan.fit_predict(scalar_df),columns = ['Clusterid_DBSCAN'])
    print(f'For eps = {n}','\n',y['Clusterid_DBSCAN'].value_counts())
```

```python
# silhouette score

sil_score = silhouette_score(scalar_df, dbscan.labels_)
print(f'For eps silhouette score = {n}','\n', sil_score)
```

```python
dbscan = DBSCAN(eps = 0.28, min_samples = 4)
dbscan.fit(scalar_df)
```

```python
df['Clusterid_DBSCAN'] = dbscan.labels_
df.head()
```

```python
df.groupby('Clusterid_DBSCAN').agg(['mean']).reset_index()
```

```python
# Plotting barplot using groupby method
```

```python
df.groupby(['Clusterid_DBSCAN']).count()['State'].plot(kind='bar')
plt.title('DBSCAN Clustering')
plt.xlabel('Clusters')
plt.ylabel('States')
```
